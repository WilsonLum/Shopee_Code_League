## Reading dataset

```python
#=========================================================================
# Read train data using pandas and first 5 rows of data
#=========================================================================

xls1    = pd.ExcelFile('data/test_dataset6_scaled.xlsx')
X_train = pd.read_excel(xls1, 'X_train_scaled')

dataset = X_train
features = ['attr_2', 'attr_3', 'subject_line_length', 'last_open_day',
       'last_login_day', 'last_checkout_day', 'open_count_last_10_days',
       'open_count_last_30_days', 'open_count_last_60_days',
       'login_count_last_10_days', 'login_count_last_30_days',
       'login_count_last_60_days', 'checkout_count_last_10_days',
       'checkout_count_last_30_days', 'checkout_count_last_60_days', 'weekday',
       'domain_@gmail.com', 'domain_@outlook.com', 'domain_@ymail.com',
       'country_code_1', 'country_code_2', 'country_code_3', 'country_code_4',
       'country_code_5', 'country_code_7']

dataset['row_id']    = 0
dataset['open_flag'] = 0

dataset.head()
```

<h3>Random Forest Classifier<h3>

```python
## We can save the trained model clf using pickle 
filename = "model/Random_Forest_trained_model_dataset6.p"
model = pickle.load(open(filename, 'rb'))
```

```python
dataset.dtypes
```

```python
row = next(dataset.iterrows())[1]
row = row[features]
row 
```

```python
row.astype(float)
row.dtypes
```

```python
row_t.astype(float64)
row_t.dtypes
```

```python
predictions = model.predict(row_t)
```

```python
predictions
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
for i, row in dataset.iterrows():
    if i%1000==0:
        print('batch:',i)
    row = row[features]
    row = pd.DataFrame(row)
    row_t =row.T
    predictions = model.predict(row_t)
    dataset['row_id'][i]    = i
    dataset['open_flag'][i] = predictions[0]
    
```

```python
dataset.columns
```

```python
dataset = dataset.drop(features, axis=1)
dataset
```

```python
dataset.to_csv('submit/Random_forest_dataset6-submission.csv', index=False)
```
