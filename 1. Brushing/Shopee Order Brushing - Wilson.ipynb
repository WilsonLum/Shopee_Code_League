```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/order_brush_order.csv')
df
```

```python
df.describe(include='all')
```

```python
df['date']=pd.to_datetime(df.event_time)
```

```python
df
```

```python
df_userid = df.groupby([pd.Grouper(key='date',freq='H'),df.userid]).size().reset_index(name='count')
```

```python
df1_userid_suspicious = df_userid[df_userid['count']>=3]
df1_userid_suspicious
```

```python
df2 = df[df['userid'].isin(df1_userid_suspicious['userid'])]
df2
```

```python
df3 = df2.groupby(df2.columns.tolist()).size().reset_index().\
    rename(columns={0:'records'})
```

```python
df3
```

```python
df4 = df3.groupby(['shopid','userid']).size().reset_index(name='count') 
df4
```

```python
df5 = df4.loc[df4['count'] >= 3].reset_index()
df5
```

```python
shopidlist = df.shopid.unique()

d = {'shopid': shopidlist, 'userid': np.zeros(len(shopidlist), dtype = "int64")}

answer = pd.DataFrame(data = d)
```

```python
shopidlist = df.shopid.unique()

d = {'shopid': shopidlist, 'userid': np.zeros(len(shopidlist), dtype = "str")}

answer = pd.DataFrame(data = d)
```

```python
answer
```

```python
for i in range(len(df5)):
    for j in range(len(answer)):
        if answer['shopid'][j] == df5['shopid'][i]:
            if answer['userid'][j] != '':
                answer['userid'][j] = answer['userid'][j] + '&' + df5['userid'][i].astype(str)
            else:
                answer['userid'][j] = (df5['userid'][i]).astype(str)
```

```python
answer
```

```python
answer.to_csv('submission.csv',index=False)
```
